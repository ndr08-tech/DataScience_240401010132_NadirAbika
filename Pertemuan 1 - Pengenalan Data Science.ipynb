```python
Data Science adalah bidang multidisiplin yang menggabungkan statistika dan matematika, ilmu komputer, serta pengetahuan domain untuk mengolah data menjadi informasi dan pengetahuan yang berguna.
```

```python
Tiga Pilar Utama Data Science
1. Statistika & Matematika → analisis data, probabilitas, regresi, dan pengujian hipotesis.
2. Ilmu Komputer → pemrograman, basis data, machine learning, dan AI.
3. Domain Knowledge → pemahaman konteks bidang tertentu seperti bisnis, kesehatan, atau teknik.
```

```python
Fenomena Big Data (4V)
- Volume: jumlah data sangat besar.
- Velocity: data dihasilkan dan diproses dengan cepat.
- Variety: data memiliki banyak bentuk (teks, gambar, video, dll.).
- Veracity: kualitas dan kebenaran data harus dijaga.
```
